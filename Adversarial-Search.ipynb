```python
# 2) Extend the minimax algorithm to handle heuristic evaluation functions.
#Submitted By: Murad Tadesse    - ID: GSR/9304/16
from math import inf

def evaluate(board):
    # Heuristic evaluation function
    patterns = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Columns
        [0, 4, 8], [2, 4, 6]  # Diagonals
    ]
    x_count = o_count = 0
    for pattern in patterns:
        if board[pattern[0]] == board[pattern[1]] == board[pattern[2]]:
            if board[pattern[0]] == 'X':
                x_count += 1
            elif board[pattern[0]] == 'O':
                o_count += 1
    return x_count - o_count  # Return the heuristic value

def minimax(board, depth, is_maximizing):
    score = evaluate(board)
    if score == 1 or score == -1 or len(empty_cells(board)) == 0 or depth == 0:
        return score

    if is_maximizing:
        best_score = -inf
        for cell in empty_cells(board):
            board[cell] = 'X'
            score = minimax(board, depth - 1, False)
            board[cell] = ' '  # Undo the move
            best_score = max(best_score, score)
        return best_score if best_score != -inf else 0  # Handle no best score found
    else:
        best_score = inf
        for cell in empty_cells(board):
            board[cell] = 'O'
            score = minimax(board, depth - 1, True)
            board[cell] = ' '  # Undo the move
            best_score = min(best_score, score)
        return best_score if best_score != inf else 0  # Handle no best score found

def empty_cells(board):
    return [i for i, cell in enumerate(board) if cell == ' ']

def best_move(board, player):
    if player == 'X':
        maximizing = True
    else:
        maximizing = False

    best_score = -inf if maximizing else inf
    best_move = None
    for cell in empty_cells(board):
        if board[cell] == ' ':
            board[cell] = player
            score = minimax(board, 5, not maximizing)
            board[cell] = ' '  # Undo the move
            if (maximizing and score > best_score) or (not maximizing and score < best_score):
                best_score = score
                best_move = cell
    return best_move

def print_board(board):
    for i in range(3):
        print(" | ".join(board[i*3:i*3+3]))
        if i < 2:
            print("---------")

# Example usage
current_player = 'X'
game_board = [' '] * 9

for _ in range(15):  # Play 5 moves, you can adjust this
    print_board(game_board)
    print(f"\nCurrent player: {current_player}")
    
    if current_player == 'X':
        move = best_move(game_board, current_player)
    else:
        move = int(input("Enter your move (0-8): "))
    
    game_board[move] = current_player

    if evaluate(game_board) in [1, -1]:
        print_board(game_board)
        print(f"\n{current_player} wins!")
        break
    elif ' ' not in game_board:
        print_board(game_board)
        print("\nIt's a tie!")
        break
    
    current_player = 'O' if current_player == 'X' else 'X'
```
